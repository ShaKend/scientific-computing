```python
import numpy as np
import matplotlib.pyplot as plt

# w + a >> new code ky gni
```

```python
x = np.array([1, 2, 3, 4, 5, 6, 7, 8])

y = np.array([2, 5, 6, 7, 9, 10, 12, 14])

A = np.vstack([x, np.ones(len(x))]).T

Y = y[:, np.newaxis]

# alpha = np.dot(A.T, A) transpose
# inverse & T:
# alpha = np.dot(np.dot(np.linalg.inv(np.dot(A.T, A)), A.T), Y)

alpha = np.linalg.pinv(A).dot(Y)
print(alpha)

# 1 2 3 4 5 6 7 8
# 1 1 1 1 1 1 1 1
# ditranspose jadi :
# 1 1
# 2 1
# 3 1
# 4 1
# 5 1
# 6 1
# 7 1
# 8 1

# plt.title("Regression")
plt.title("Least Square Regression")
plt.plot(x, y, 'or') 
plt.plot(x, alpha[0] * x + alpha[1])
# plt.plot(x, alpha[0] * x + alpha[1], r)
# m = 1.5833
# b = 1
# plt.plot(x, m * x + b)

# o, or, ob, -, .g, without parameter jg bsa >> (x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['data', 'regression'])
plt.show()

# untuk ngebuat garis harus >> y = mx + b
```
