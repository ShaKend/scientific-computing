```python
import numpy as np
import matplotlib.pyplot as plt;
```

```python
x = np.array([2, 4, 6, 7, 8])
y = np.array([12, 14, 15, 17, 19])

A = np.vstack([x,np.ones(len(x))]).T;
B = y[:, np.newaxis];

alpha = np.linalg.pinv(A).dot(B);
print(alpha);

plt.title("least square regression");
plt.plot(x, y, "or");
plt.plot(x, alpha[0] * x + alpha[1]);

plt.xlabel('x');
plt.ylabel('y');
plt.legend('data', 'regresi');

plt.show();
```
