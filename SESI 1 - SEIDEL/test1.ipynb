```python
import numpy as np;
```

```python
# MENERIMANYA ARRAY BKN YG LAIN YA!!!!!!
def diagonal(x):
    x = np.array(x);
    diag = np.diag(np.abs(x));
    print(f"Diagonal: {diag}");
    
    off_diag = np.sum(np.abs(x), axis=1) - diag;
    
    if(np.all(diag > off_diag)):
        return True;
    else:
        return False;
    
```

```python
arr1 = [
    [9, 4, 2],
    [3, 8, 1],
    [2, 1, 7]
]
# syarat untuk return true
# diagonal matrix nya dikurang jumlah matrix sisanya(lakukan per row)

print(diagonal(arr1))
```

```python
def gauss(x, y, epsilon = 0.1, max_t = 15):
    x = np.array(x)
    y = np.array(y)
    
    diag = np.diag(x);
    
    x = -x;
    
    #isi diagonal menjadi 0
    np.fill_diagonal(x, 0);
    
    #inisiasi value x
    old = np.zeros(x[0].shape)
    
    for i in range(max_t):
        new = np.array(old);
        
        for idx, row in enumerate(x):
            print(f"index: {idx}, row: {row}");
            print();
            
            new[idx] = (y[idx] + np.dot(row, new)) / diag[idx];
            
            print(f"matrix ke-{idx}: {new}");
    
    print(f"Iterasi-{i}: {new}")
    
    dx = np.sqrt(np.dot(new - old, new - old))
    
    if(dx < epsilon):
        print("converged");
        
    old = new;
    print("not converged");
    return;
```

```python
Xa = [
    [
        [4, 6, 3],
        [2, 1, 4],
        [7, 3, 1]
    ]
]
Ya = [
    [2, 1, 2],
    [6, 3, 8],
]

for i, (X1, Y1) in enumerate(zip(Xa, Ya)):
    print(f"persamaan {i - 1}: ")
    gauss(X1, Y1)
```
