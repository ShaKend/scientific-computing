```python
import numpy as np
```

```python
def isDiagonallyDominant(x):
    # ubah list jadi array
    x = np.array(x)

    # ambil nilai diagonal + absolut
    diag = np.diag(np.abs(x))
    print(f'Diagonal Value: {diag}')

    # ambil nilai selain diagonal
    # axis = 0 -> column
    # axis = 1 -> row
    off_diag = np.sum(np.abs(x), axis=1) - diag # objek

    if np.all(diag > off_diag):
        return True
    else:
        return False
```

```python
Xs = [
    [10, 4, 5],
    [1, 6, 1],
    [1, 2, 5]
]

print(isDiagonallyDominant(Xs))
```

```python
def gauss_seidel(x, y, epsilon= 0.01, max_t = 15):  #max_t = max iterationnya

    # ubah list jadi array
    x = np.array(x)
    y = np.array(y)

    # ambil diagonal
    diag = np.diag(x)

    # negasi dari x nya
    x = -x

    # isi diagonal menjadi 0
    np.fill_diagonal(x, 0)

    # inisialisasi value x nya
    # cnth : x => 3x3
    # mtx_old = [0, 0, 0]
    #            x1 x2 x3 
    mtx_old = np.zeros(x[0].shape)

    for i in range(max_t):
            
        mtx_new = np.array(mtx_old)

        # enumarate -> untuk dapetin index dan value

        for idx, row in enumerate(x):
            print(f'idx: {row}, row: {row}')
            print()

            mtx_new[idx] = (y[idx] + np.dot(row, mtx_new)) / diag[idx]

            print(f'Matrix new-{idx}: {mtx_new}')
        
    print(f'Iterasi-{i}: {mtx_new}')

    #                     dikuadrat
    dx = np.sqrt(np.dot(mtx_new - mtx_old, mtx_new - mtx_old))

    # cek distance lebih kecil dari threshold
    if dx < epsilon:
        print('converged')
        return
        
    # majuin (value) mtx_old ke terasi sekarang
    mtx_old = mtx_new

    # kondisi kalau sampai batas iterasi masih belum lebih kecil dari threshold 
    print("Not Converged")
    return
```

```python
Xm = [
    [
        [10, 4, 5],
        [1, 6, 1],
        [1, 2, 5]
    ],
    [
        [9, 3, 2, 1],
        [2, 8, 2, 3],
        [2, 1, 4, 2],
        [4, 3, 1, 10]
    ]
]

Ym = [
    [8, 5, 10],
    [10, 20, 21, 8]
]

for i, (Xa, Ys) in enumerate(zip(Xm, Ym)):

    print(f'Persamaan:-{i + 1}:')
    gauss_seidel(Xa, Ys)
```
